#Praktikum 01: try...except Dasar dalam Metode Kelas

```python
class DataStore:
  def __init__(self, nama_store="Default Store"):
    # Dictionary internal untuk menyimpan data (anggap sebagai 'protected')
    self._data = {}
    self.nama = nama_store
    print(f"DataStore '{self.nama} dibuat.")

  def tambah_data(self, key, value):
    """Menambahkan data ke dalam store."""
    print(f"[{self.nama}] Menambahkan data: '{key}' -> {value}")
    self._data[key] = value
    print(" -> Data berhasil ditambahkan.")

  def get_data(self, key):
    """
    Mencoba mendapatkan data berdasarkan key.
    Menangani KeyError jika key tidak ada.
    """
    print(f"[{self.nama}] Mencoba mengambil data dengan key: '{key}'")
    try:
      #Baris ini berpotensi menimbulkan KeyError
      nilai = self._data[key]
      print(f" -> Data ditemukan : {nilai}")
      return nilai
    except KeyError:
      # Blok ini dijalankan jika KeyError terjadi di blok try
      print(f" -> ERROR: Key '{key}' tidak ditemukan di DataStore'{self.nama}'.")
      # Mengembalikan None atau nilai default lain sebagai indikasi gagal
      return None

# --- Kode Utama ---
if __name__ == "__main__":
  # Membuat instance DataStore
  store_produk = DataStore("Toko Produk")

  # Menambahkan beberapa data
  store_produk.tambah_data("Buku Tulis", 15000)
  store_produk.tambah_data("Pensil 2B", 2000)

  # Mengambil data yang ada (seharusnya berhasil)
  print("\nMengambil data yang ADA:")
  harga_buku = store_produk.get_data("Buku Tulis")
  if harga_buku is not None:
    print(f"Harga Buku Tulis adalah: {harga_buku}")

  # Mengambil data yang TIDAK ada (seharusnya memicu except)
  print("\nMengambil data yang TIDAK ADA:")
  harga_penggaris = store_produk.get_data("Penggaris")
  if harga_penggaris is None:
    print("Harga penggaris tidak ditemukan ()")
```

#Praktikum 02: Menagani Beberapa Exception Spesifik

```python
class KalkulatorAngka:
    def __init__(self):
        print("Kalkulator Angka siap.")

    def bagi(self, angka1, angka2):
        """
        Membagi angka1 dengan angka2.
        Menangani potensi ValueError (jika input bukan angka)
        dan ZeroDivisionError (jika pembagi nol) secara terpisah.
        """

        print(f"\nMencoba membagi '{angka1}' dengan '{angka2}'...")

        try:
            # Konversi input ke float, bisa memicu ValueError
            num1 = float(angka1)
            num2 = float(angka2)

            # Operasi pembagian, bisa memicu ZeroDivisionError
            hasil = num1 / num2

            print(f"  -> Hasil pembagian: {hasil}")
            return hasil

        # Menangkap ValueError secara spesifik
        except ValueError:
            print("  -> ERROR: Input tidak valid! Pastikan kedua input adalah angka.")
            return None

        # Menangkap ZeroDivisionError secara spesifik
        except ZeroDivisionError:
            print("  -> ERROR: Tidak bisa melakukan pembagian dengan nol.")
            return None

        # Menangkap exception lain yang mungkin tidak terduga (opsional)
        except Exception as e:
            print(f"  -> ERROR Lainnya: Terjadi kesalahan - {e}")
            return None


# --- Kode Utama ---
if __name__ == "__main__":
    calc = KalkulatorAngka()

    print("=== Tes Kalkulator ===")

    hasil1 = calc.bagi(100, 5)          # Operasi sukses
    hasil2 = calc.bagi(20, 0)           # Memicu ZeroDivisionError
    hasil3 = calc.bagi("limapuluh", 10) # Memicu ValueError saat konversi num1
    hasil4 = calc.bagi(50, "sepuluh")   # Memicu ValueError saat konversi num2

    print("\n--- Ringkasan Hasil ---")
    print(f"Hasil 1 (100/5): {hasil1}")
    print(f"Hasil 2 (20/0): {hasil2}")
    print(f"Hasil 3 ('limapuluh'/10): {hasil3}")
    print(f"Hasil 4 (50/'sepuluh'): {hasil4}")
```

#Praktikum 03: Menggunakan else dan finally dalam Metode

```python
import os # Diperlukan untuk menghapus file dummy nanti

class FileProcessor:
    def __init__(self, nama_file):
        self.nama_file = nama_file
        self._file_handle = None # Untuk menyimpan objek file saat dibuka
        print(f"\n-> FileProcessor dibuat untuk file: '{self.nama_file}'")

    def baca_dan_proses(self):
        """Membuka, membaca, dan 'memproses' file, dengan cleanup di finally."""
        print(f"[{self.nama_file}] Mencoba memproses...")
        try:
            # 1. Mencoba membuka file untuk dibaca
            print(f"  [{self.nama_file}] Membuka file...")
            self._file_handle = open(self.nama_file, 'r', encoding='utf-8') # Mode 'r' untuk baca

            # 2. Jika buka berhasil, coba baca isinya
            print(f"  [{self.nama_file}] Membaca isi file...")
            isi = self._file_handle.read()
            print(f"  -> File berhasil dibaca.")

            # 3. Simulasi pemrosesan isi
            print(f"  [{self.nama_file}] Memproses isi...")
            panjang_isi = len(isi)
            print(f"  -> Panjang isi: {panjang_isi} karakter.")
            # Simulasi: Anggap pemrosesan selalu berhasil jika baca berhasil

        except FileNotFoundError:
            # 4a. Blok ini jalan jika open() gagal karena file tidak ada
            print(f"  -> ERROR: File '{self.nama_file}' tidak ditemukan.")
        except UnicodeDecodeError:
            # 4b. Contoh error lain saat membaca file non-teks
            print(f"  -> ERROR: Gagal membaca '{self.nama_file}' sebagai teks (mungkin file biner?).")
        except Exception as e:
            # 4c. Menangkap error tak terduga lainnya saat try
            print(f"  -> ERROR Lainnya saat memproses file: {type(e).__name__} - {e}")
        else:
            # 5. Blok ini jalan HANYA JIKA try selesai tanpa exception
            print(f"  -> PEMROSESAN BERHASIL: Tidak ada error terjadi pada '{self.nama_file}'.")
        finally:
            # 6. Blok ini SELALU jalan setelah try/except/else selesai
            print(f"  [{self.nama_file}] Blok Finally: Melakukan cleanup...")
            if self._file_handle and not self._file_handle.closed:
                print(f"  Menutup file '{self.nama_file}'...")
                self._file_handle.close()
                print(f"  -> File berhasil ditutup.")
            else:
                # Ini akan terjadi jika file gagal dibuka (FileNotFoundError)
                # atau jika sudah ditutup karena error lain di try
                print(f"  -> File tidak perlu ditutup (belum dibuka atau sudah ditutup).")

        print(f"[{self.nama_file}] Selesai mencoba proses.")
        print("-" * 30)


# --- Kode Utama ---
if __name__ == "__main__":
    NAMA_FILE_SUKSES = "data_penting.txt"
    NAMA_FILE_GAGAL = "file_yang_tidak_ada.log"

    # Membuat file dummy untuk skenario sukses
    try:
        with open(NAMA_FILE_SUKSES, "w", encoding='utf-8') as f:
            f.write("Ini adalah baris pertama.\n")
            f.write("Baris kedua berisi data penting.\n")
        print(f"File dummy '{NAMA_FILE_SUKSES}' berhasil dibuat.")
    except IOError as e:
        print(f"Gagal membuat file dummy '{NAMA_FILE_SUKSES}': {e}")

    # Skenario 1: File ada dan proses berhasil
    print("\n--- Skenario 1: Proses Sukses ---")
    processor1 = FileProcessor(NAMA_FILE_SUKSES)
    processor1.baca_dan_proses()

    # Skenario 2: File tidak ada
    print("\n--- Skenario 2: File Tidak Ditemukan ---")
    processor2 = FileProcessor(NAMA_FILE_GAGAL)
    processor2.baca_dan_proses()

    # Cleanup file dummy
    try:
        if os.path.exists(NAMA_FILE_SUKSES):
            os.remove(NAMA_FILE_SUKSES)
            print(f"\nFile dummy '{NAMA_FILE_SUKSES}' dihapus.")
    except OSError as e:
        print(f"\nGagal menghapus file dummy '{NAMA_FILE_SUKSES}': {e}")
```

#Praktikum 04: Mengakses Objek Exception (as e)

```python
class KalkulatorAngka:
    def __init__(self):
        print("Kalkulator Angka siap.")

    def bagi(self, angka1, angka2):
        """
        Membagi angka1 dengan angka2.
        Menangkap objek exception untuk menampilkan detail error.
        """
        print(f"\nMencoba membagi '{angka1}' dengan '{angka2}'...")
        try:
            num1 = float(angka1)
            num2 = float(angka2)
            if num2 == 0:
                # Bisa juga raise ZeroDivisionError di sini jika mau
                raise ZeroDivisionError("Pembagi tidak boleh nol secara eksplisit.")
            hasil = num1 / num2
            print(f"  -> Hasil pembagian: {hasil}")
            return hasil

        # Menangkap ValueError dan menyimpannya di variabel 'e_val'
        except ValueError as e_val:
            print(f"  -> ERROR Konversi: Input tidak valid.")
            # Menampilkan representasi string dari objek exception e_val
            print(f"     Detail Error (ValueError): {e_val}")
            return None

        # Menangkap ZeroDivisionError dan menyimpannya di variabel 'e_zero'
        except ZeroDivisionError as e_zero:
            print(f"  -> ERROR Pembagian: Tidak dapat membagi dengan nol.")
            # Menampilkan representasi string dari objek exception e_zero
            print(f"     Detail Error (ZeroDivisionError): {e_zero}")
            return None

        # Menangkap exception lain dan menyimpannya di variabel 'e'
        except Exception as e:
            print(f"  -> ERROR Lainnya: Terjadi kesalahan tak terduga.")
            # Menampilkan representasi string dari objek exception e
            print(f"     Detail Error (Lainnya): {type(e).__name__} - {e}")
            return None


# --- Kode Utama ---
if __name__ == "__main__":
    calc = KalkulatorAngka()

    print("--- Tes Kalkulator ---")
    hasil1 = calc.bagi(100, 5)           # Sukses
    hasil2 = calc.bagi(20, 0)            # Memicu ZeroDivisionError
    hasil3 = calc.bagi("limapuluh", 10)  # Memicu ValueError
    hasil4 = calc.bagi(10, [])           # Memicu TypeError (ditangkap except Exception)

    print("\n--- Ringkasan Hasil ---")
    print(f"Hasil 1 (100/5): {hasil1}")
    print(f"Hasil 2 (20/0): {hasil2}")
    print(f"Hasil 3 ('limapuluh'/10): {hasil3}")
    print(f"Hasil 4 (10/[]): {hasil4}")
```

#Praktikum 05: Membangkitkan Exception (raise) dalam Metode

```python
# Helper untuk format Rupiah (opsional, bisa disalin dari praktikum sebelumnya)
import locale
try:
    locale.setlocale(locale.LC_ALL, 'id_ID.UTF-8')
except locale.Error:
    print("Locale id_ID tidak tersedia.")

def format_rupiah(angka):
    try:
        return locale.currency(angka, grouping=True, symbol='Rp ')
    except (NameError, ValueError): # Catch NameError or ValueError (from locale.currency if locale is not set)
        return f"Rp {angka:,.2f}"


class PengelolaAkun:
    def __init__(self, nomor_akun, saldo_awal=0):
        print(f"\nMencoba membuat akun {nomor_akun}...")
        if not isinstance(saldo_awal, (int, float)) or saldo_awal < 0:
            # 1. Raise ValueError di __init__ jika saldo awal tidak valid
            raise ValueError(f"Saldo awal ({saldo_awal}) tidak valid. Harus angka non-negatif.")
        self.nomor_akun = nomor_akun
        self._saldo = float(saldo_awal)
        print(f"  -> Akun {self.nomor_akun} berhasil dibuat dengan saldo: {format_rupiah(self._saldo)}")

    def get_saldo(self):
        """Mengembalikan saldo saat ini."""
        # print(f"[Info Akun {self.nomor_akun}] Saldo saat ini: {format_rupiah(self._saldo)}")
        return self._saldo

    def tarik_tunai(self, jumlah):
        """Menarik tunai dengan validasi. Me-raise exception jika gagal."""
        print(f"\n[{self.nomor_akun}] Mencoba tarik tunai: {format_rupiah(jumlah)}")

        # 2. Validasi jumlah penarikan
        if not isinstance(jumlah, (int, float)) or jumlah <= 0:
            # Raise ValueError jika jumlah tidak valid
            raise ValueError(f"Jumlah penarikan ({jumlah}) tidak valid. Harus angka positif.")

        # 3. Validasi saldo cukup
        if jumlah > self._saldo:
            # Raise RuntimeError (atau exception kustom) jika saldo tidak cukup
            raise RuntimeError(f"Saldo tidak mencukupi! Saldo: {format_rupiah(self._saldo)}, Diminta: {format_rupiah(jumlah)}")

        # Jika semua validasi lolos, lakukan penarikan
        self._saldo -= jumlah
        print(f"  -> Penarikan {format_rupiah(jumlah)} berhasil.")
        print(f"  -> Saldo sekarang: {format_rupiah(self._saldo)}")
        return True # Mengindikasikan sukses


# --- Kode Utama ---
if __name__ == "__main__":
    # Skenario 1: Gagal membuat akun karena saldo awal tidak valid
    print("\n--- Skenario 1: Saldo Awal Tidak Valid ---")
    try:
        akun_gagal = PengelolaAkun("ACC001", -50000)
    except ValueError as e:
        print(f"  ERROR DITANGKAP saat pembuatan akun: {e}")

    # Skenario 2: Membuat akun valid, lalu coba transaksi
    print("\n--- Skenario 2: Akun Valid dan Transaksi ---")
    akun_sukses = None # Inisialisasi akun_sukses untuk memastikan variabel selalu ada dalam scope
    try:
        akun_sukses = PengelolaAkun("ACC002", 100000)

        # Transaksi 1: Sukses
        if akun_sukses: # Pastikan akun_sukses berhasil dibuat sebelum mencoba transaksi
            akun_sukses.tarik_tunai(30000)

            # Transaksi 2: Gagal karena jumlah tidak valid (ValueError)
            print("\n  Mencoba tarik jumlah negatif...")
            akun_sukses.tarik_tunai(-10000)

    except ValueError as e:
        print(f"  ERROR DITANGKAP (ValueError): {e}")
    except RuntimeError as e:
        print(f"  ERROR DITANGKAP (RuntimeError): {e}")

    # Skenario 3: Melanjutkan akun valid, gagal karena saldo tidak cukup
    print("\n--- Skenario 3: Saldo Tidak Cukup ---")
    if akun_sukses: # Hanya lanjutkan jika akun_sukses berhasil dibuat
        try:
            print(f"  Saldo akun ACC002 saat ini: {format_rupiah(akun_sukses.get_saldo())}")
            print("  Mencoba tarik jumlah > saldo...")
            akun_sukses.tarik_tunai(80000)  # Saldo hanya sisa 70000

        except ValueError as e:
            print(f"  ERROR DITANGKAP (ValueError): {e}")
        except RuntimeError as e:
            print(f"  ERROR DITANGKAP (RuntimeError): {e}")  # Ini yang diharapkan tertangkap

        print(f"\nSaldo akhir akun ACC002: {format_rupiah(akun_sukses.get_saldo())}")
    else:
        print("  Akun ACC002 tidak berhasil dibuat, tidak dapat melanjutkan Skenario 3.")
```

#Praktikum 06: Kombinasi Menangkap dan Membangkitkan Exception

```python
# 1. Definisikan Custom Exception (mewarisi dari Exception)
class KonfigurasiError(Exception):
    """Exception khusus untuk menandakan error terkait konfigurasi aplikasi."""
    def __init__(self, message, penyebab_asli=None):
        super().__init__(message)
        self.penyebab_asli = penyebab_asli  # Menyimpan error asli jika ada

    def __str__(self):
        if self.penyebab_asli:
            return f"{super().__str__()} (Disebabkan oleh: {type(self.penyebab_asli).__name__})"
        return super().__str__()


# Kelas yang menggunakan exception handling gabungan
class PengelolaSettings:
    def __init__(self, file_config):
        self._config = {}
        self.file_config = file_config
        # Panggil metode internal yang berpotensi raise KonfigurasiError
        try:
            self._load_config()
            print(f"  -> PengelolaSettings berhasil dibuat untuk '{self.file_config}'.")
        except KonfigurasiError as e:
            print(f"  -> GAGAL membuat PengelolaSettings: {e}")
            # Bisa jadi kita ingin raise lagi agar pembuat objek tahu ada masalah serius
            # raise e
            # Atau set state default / kosong
            self._config = {}
            print(f"     -> Menggunakan konfigurasi default/kosong.")

    def _load_config(self):
        """Metode internal untuk load config. Bisa raise KonfigurasiError."""
        print(f"\n  [{self.file_config}] Mencoba load konfigurasi...")
        try:
            # --- Potensi FileNotFoundError ---
            print(f"    Membuka file '{self.file_config}'...")
            # Simulasi: anggap hanya file 'config.txt' yang valid untuk dibaca
            if self.file_config != 'config.txt':
                # Jika file tidak ada, kita raise FileNotFoundError secara manual
                # (dalam aplikasi nyata, ini akan otomatis dari open())
                raise FileNotFoundError(f"File '{self.file_config}' tidak ada.")

            # --- Potensi Error Isi File (KeyError, ValueError) ---
            print(f"    Membaca dan validasi isi '{self.file_config}'...")
            # Simulasi data raw (bisa dari json.load(f), dll)
            # data_raw = {"host": "192.168.1.1", "port": "8080"}  # Port benar
            data_raw = {"host": "192.168.1.1", "port": "xyz"}  # Port salah tipe
            # data_raw = {"host": "192.168.1.1"}  # Key 'port' hilang

            # Cek key 'port'
            port_str = data_raw['port']  # Bisa raise KeyError

            # Cek tipe 'port'
            port_int = int(port_str)  # Bisa raise ValueError

            # Jika lolos, simpan
            self._config['port'] = port_int
            self._config['host'] = data_raw.get('host', 'localhost')
            print(f"     -> Konfigurasi berhasil di-load dan divalidasi.")

        # --- Blok Except ---
        except FileNotFoundError as e_fnf:
            print(f"    ERROR Bawaan: File tidak ditemukan.")
            # Re-raise sebagai KonfigurasiError
            raise KonfigurasiError(f"File konfigurasi penting '{self.file_config}' hilang.",
                                   penyebab_asli=e_fnf) from e_fnf
        except KeyError as e_key:
            print(f"    ERROR Bawaan: Kunci konfigurasi hilang.")
            # Re-raise sebagai KonfigurasiError
            raise KonfigurasiError(f"Kunci '{e_key}' wajib ada dalam file konfigurasi.",
                                   penyebab_asli=e_key) from e_key
        except ValueError as e_val:
            print(f"    ERROR Bawaan: Nilai konfigurasi tidak valid.")
            # Re-raise sebagai KonfigurasiError
            raise KonfigurasiError(f"Nilai untuk 'port' ('{port_str}') harus integer.",
                                   penyebab_asli=e_val) from e_val
        except Exception as e_lain:
            print(f"    ERROR Bawaan Tak Terduga: {type(e_lain).__name__}")
            # Re-raise sebagai KonfigurasiError umum
            raise KonfigurasiError("Terjadi error tak terduga saat memuat konfigurasi.",
                                   penyebab_asli=e_lain) from e_lain

    def get_setting(self, key, default=None):
        """Mendapatkan nilai setting."""
        # print(f"Mendapatkan setting '{key}'...")
        return self._config.get(key, default)


# --- Kode Utama ---
if __name__ == "__main__":
    # Skenario 1: File tidak ada -> Menghasilkan KonfigurasiError
    print("\n--- Skenario 1: File Salah ---")
    # Error sudah ditangani di dalam __init__ berdasarkan contoh ini
    settings1 = PengelolaSettings("setting.cfg")
    # Jika error di-raise ulang oleh init, kode di bawah tidak jalan
    print(f"  Port dari settings1: {settings1.get_setting('port', 'Belum terload')}")

    # Skenario 2: File ada, tapi data port salah -> Menghasilkan KonfigurasiError
    print("\n--- Skenario 2: Data Port Salah ---")
    # (Anggap 'config.txt' ada tapi isinya menyebabkan ValueError di _load_config)
    settings2 = PengelolaSettings("config.txt")
    print(f"  Port dari settings2: {settings2.get_setting('port', 'Belum terload')}")

    # Skenario 3: (Jika simulasi di _load_config diubah jadi benar)
    print("\n--- Skenario 3: Sukses ---")
    settings3 = PengelolaSettings("config.txt")  # Gunakan data_raw yang benar
    print(f"  Port dari settings3: {settings3.get_setting('port')}")
    print(f"  Host dari settings3: {settings3.get_setting('host')}")
```

#Penugasan

```python
class InventarisBarang:
    def __init__(self):
        self._items = {}  # Format: {kode_barang: jumlah_stok}
        print("Inventaris Barang siap digunakan.\n")

    # ------------------------------------------------------------------ #
    # b) tambah_barang
    # ------------------------------------------------------------------ #
    def tambah_barang(self, kode, jumlah):
        """Menambahkan jumlah stok untuk kode barang tertentu (hanya barang baru)."""
        print(f"[tambah_barang] Mencoba menambahkan '{kode}' sebanyak {jumlah}...")

        # Validasi jumlah harus integer positif
        try:
            jumlah = int(jumlah)
            if jumlah <= 0:
                raise ValueError
        except ValueError:
            print(f"  -> ERROR: Jumlah '{jumlah}' tidak valid. Harus integer positif.")
            return

        # Cek apakah kode sudah ada (tambah hanya untuk barang baru)
        if kode in self._items:
            raise KeyError(f"Error: Kode barang '{kode}' sudah ada.")

        # Jumlah valid dan kode belum ada -> tambahkan
        self._items[kode] = jumlah
        print(f"  -> SUKSES: Barang '{kode}' ditambahkan dengan stok {jumlah}.")

    # ------------------------------------------------------------------ #
    # c) ambil_barang
    # ------------------------------------------------------------------ #
    def ambil_barang(self, kode, jumlah):
        """Mengurangi jumlah stok untuk kode barang tertentu."""
        print(f"[ambil_barang] Mencoba mengambil '{kode}' sebanyak {jumlah}...")

        # Validasi jumlah harus integer positif
        try:
            jumlah = int(jumlah)
            if jumlah <= 0:
                raise ValueError
        except ValueError:
            print(f"  -> ERROR: Jumlah '{jumlah}' tidak valid. Harus integer positif.")
            return

        # Cek apakah kode barang ada
        try:
            stok_sekarang = self._items[kode]  # Bisa raise KeyError
        except KeyError:
            print(f"  -> ERROR: Kode barang '{kode}' tidak ditemukan dalam inventaris.")
            return

        # Cek apakah stok mencukupi
        if jumlah > stok_sekarang:
            raise ValueError(
                f"Error: Stok barang '{kode}' tidak mencukupi "
                f"(tersedia: {stok_sekarang}, diminta: {jumlah})."
            )

        # Kode ada dan stok mencukupi -> kurangi stok
        self._items[kode] -= jumlah
        print(f"  -> SUKSES: Mengambil {jumlah} dari '{kode}'. Sisa stok: {self._items[kode]}.")

    # ------------------------------------------------------------------ #
    # d) cek_stok
    # ------------------------------------------------------------------ #
    def cek_stok(self, kode):
        """Memeriksa dan mencetak stok untuk kode barang tertentu."""
        print(f"[cek_stok] Memeriksa stok untuk '{kode}'...")
        try:
            stok = self._items[kode]  # Bisa raise KeyError
        except KeyError:
            print(f"  -> ERROR: Barang '{kode}' tidak ditemukan.")
        else:
            # Hanya dijalankan jika try berhasil (tidak ada exception)
            print(f"  -> Stok '{kode}': {stok} unit.")
        finally:
            # Selalu dijalankan
            print(f"  -> Pengecekan stok untuk '{kode}' telah selesai dilakukan.")


# ------------------------------------------------------------------ #
# e) Kode Utama
# ------------------------------------------------------------------ #
if __name__ == "__main__":
    inventaris = InventarisBarang()

    # --- Operasi tambah_barang ---
    print("=" * 50)
    print("OPERASI TAMBAH BARANG")
    print("=" * 50)

    inventaris.tambah_barang("A001", 50)       # Sukses
    inventaris.tambah_barang("B002", 30)       # Sukses
    inventaris.tambah_barang("C003", 20)       # Sukses
    # Jumlah valid tapi kode sudah ada -> raise KeyError, tangkap di kode utama
    try:
        inventaris.tambah_barang("A001", 10)
    except KeyError as e:
        print(f"  [Kode Utama] KeyError DITANGKAP: {e}")

    try:
        inventaris.tambah_barang("B002", 5)
    except KeyError as e:
        print(f"  [Kode Utama] KeyError DITANGKAP: {e}")

    inventaris.tambah_barang("D004", -5)       # Jumlah tidak valid
    inventaris.tambah_barang("E005", "banyak") # Jumlah tidak valid (bukan integer)

    # --- Operasi ambil_barang ---
    print()
    print("=" * 50)
    print("OPERASI AMBIL BARANG")
    print("=" * 50)

    # Kode tidak ada -> KeyError dari dalam metode (ditangani di dalam)
    inventaris.ambil_barang("Z999", 5)

    # Jumlah tidak valid -> ValueError dari dalam metode (ditangani di dalam)
    inventaris.ambil_barang("A001", -10)

    # Stok tidak cukup -> ValueError yang di-raise manual
    try:
        inventaris.ambil_barang("C003", 100)   # Stok hanya 20
    except ValueError as e:
        print(f"  [Kode Utama] ValueError DITANGKAP: {e}")

    # Pengambilan sukses
    try:
        inventaris.ambil_barang("A001", 15)    # Stok A001 = 50, ambil 15
        inventaris.ambil_barang("B002", 30)    # Stok B002 = 30, ambil semua
    except (KeyError, ValueError) as e:
        print(f"  [Kode Utama] Exception DITANGKAP: {e}")

    # --- Operasi cek_stok ---
    print()
    print("=" * 50)
    print("OPERASI CEK STOK")
    print("=" * 50)

    inventaris.cek_stok("A001")   # Ada -> tampil stok + finally
    print()
    inventaris.cek_stok("B002")   # Ada (stok 0) -> tampil stok + finally
    print()
    inventaris.cek_stok("Z999")   # Tidak ada -> KeyError + finally
    print()
    inventaris.cek_stok("C003")   # Ada -> tampil stok + finally
```
